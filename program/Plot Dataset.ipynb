```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
data = pd.read_excel("D:\PROPOSAL SKRIPSI\PLOTTING AKURASI\corr_mlbb.xlsx")
df = pd.DataFrame(data)
bar = sns.barplot(x=df['Classifier'],y=df['Akurasi'], hue=df['Classifier'],data = df, dodge=False)
bar.set_title('Akurasi Klasifikasi Hasil Seleksi Fitur Algoritma Pearson Correlation Coefficient MLBB Dataset', fontdict= { 'fontsize': 24, 'fontweight':'bold'})
sns.set(rc = {'figure.figsize':(20,11),'axes.facecolor':'white', 'figure.facecolor':'white'}, font_scale=2)
bar.set(xlabel='Classifier', ylabel='Akurasi')
plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0., title="Classifier")
```
